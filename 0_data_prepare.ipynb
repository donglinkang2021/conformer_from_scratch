# 准备数据

## transcript

```python
def load_text(text_path="./data/aishell/transcript/aishell_transcript_v0.8.txt"):
    with open(text_path, "r") as f:
        corpus = f.read().strip("\n")
    return corpus

corpus = load_text()
```

## 简单查看path和label

```python
import os
for line in corpus.split("\n"):
    wav_path, text = line.split(" ", 1)
    split_dir = wav_path[6:11]
    if split_dir in ["S0002", "S0003"]:
        wav_path = os.path.join("./data/aishell/wav/train/", wav_path[6:11], wav_path + ".wav")
    elif split_dir in ["S0004"]:
        wav_path = os.path.join("./data/aishell/wav/dev/", wav_path[6:11], wav_path + ".wav")
    else :
        continue
    text = text.replace(" ", "")
    print(wav_path, text)
```

## 生成txt文件

```python
for line in corpus.split("\n"):
    wav_path, text = line.split(" ", 1)
    split_dir = wav_path[6:11]
    if split_dir in ["S0002", "S0003"]:
        wav_path = os.path.join("./data/aishell/wav/train/", wav_path[6:11], wav_path + ".wav")
    elif split_dir in ["S0004"]:
        wav_path = os.path.join("./data/aishell/wav/dev/", wav_path[6:11], wav_path + ".wav")
    else :
        continue
    text = text.replace(" ", "")
    with open(wav_path.replace(".wav", ".txt"), "w") as f:
        f.write(text)
    print(wav_path, text)
```

## 生成vocab文件

```python
from utils.tokenizer import build_vocab
vocab = build_vocab(data_path=f'./data/aishell/transcript/aishell_transcript_v0.8.txt')
vocab.save('./data/aishell/vocab.txt')
```

```python
for line in corpus.split('\n')[:5]:
    wav_path, text = line.split(" ", 1)
    text = text.replace(" ", "")
    print(f"text: {text}")
    print(f"tokenized_text: {vocab.to_indices(text)}")
```

## 生成数据列表文件

```python
import torch
for line in corpus.split("\n"):
    wav_path, text = line.split(" ", 1)
    split_dir = wav_path[6:11]
    if split_dir in ["S0002", "S0003"]:
        wav_path = os.path.join("./data/aishell/wav/train/", wav_path[6:11], wav_path + ".wav")
    elif split_dir in ["S0004"]:
        wav_path = os.path.join("./data/aishell/wav/dev/", wav_path[6:11], wav_path + ".wav")
    else :
        continue
    text = text.replace(" ", "")
    tokenized_text_tensor = torch.LongTensor(vocab.to_indices(text))
    torch.save(tokenized_text_tensor, wav_path.replace(".wav", ".pth"))
    print(wav_path, text)
```
